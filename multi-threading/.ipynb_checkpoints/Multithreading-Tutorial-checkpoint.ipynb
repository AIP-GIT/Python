```python
'''
Sequential programming
'''
import threading as t
import time
def display():
    for i in range(10):
        time.sleep(1)
        print(i,'Child Thread','\n')

        
display()
for i in range(10):
    time.sleep(1)
    print(i,'Main Thread','\n')
```

```python
'''
Method-1
Multithreading programming
Sequence is not guranteed
Performance benefit
'''
import threading as t
import time
def display():
    for i in range(10):
        time.sleep(1)
        print(i,'Child Thread','\n')

        
t1 = t.Thread(target=display).start()
for i in range(10):
    time.sleep(1)
    print(i,'Main Thread','\n')
```

```python
'''
Method-1.1
'''
import threading as t
import time
class Test:
    def display(self):
        for i in range(10):
            time.sleep(1)
            print(i,'Child Thread','\n')
            
obj1 = Test()
t1 = t.Thread(target=obj1.display).start()
for i in range(10):
    time.sleep(1)
    print(i,'Main Thread','\n')
```

```python
'''
Method-2
thread by extending thread class - recommended approach
extend and overwrite run method
'''
import threading as t
import time
class Mythread(t.Thread):
    def run(self):
        for i in range(10):
            time.sleep(1)
            print(i,'Child Thread','\n')
            
t1 = Mythread().start()
for i in range(10):
    time.sleep(1)
    print(i,'Main Thread','\n')
```

```python
'''
time taken without thread
'''

import time
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double : ',2*n)

def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Square : ',n*n)

numbers = [1,2,3,4,5,6]
begintime=time.time()
doubles(numbers)
squares(numbers)
endtime = time.time()
print('Total time taken : ',endtime-begintime)
```

```python
'''
showing performance benifit with threads
Not correct Main thread not waiting for child thread to complete
'''
import threading as t
import time
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double : ',2*n)

def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Square : ',n*n)

numbers = [1,2,3,4,5,6]
begintime=time.time()
t1 = t.Thread(target=doubles,args=(numbers,)) #notice the comma
t2 = t.Thread(target=squares,args=(numbers,)) #notice the comma
t1.start()
t2.start()

endtime = time.time()
print('Total time taken : ',endtime-begintime)
```

```python
'''
showing performance benifit with threads
correct Main thread waiting for child thread to complete
'''
import time
import threading as t
def doubles(numbers):
    for n in numbers:
        time.sleep(1)
        print('Double : ',2*n)

def squares(numbers):
    for n in numbers:
        time.sleep(1)
        print('Square : ',n*n)

numbers = [1,2,3,4,5,6]
begintime=time.time()
t1 = t.Thread(target=doubles,args=(numbers,)) #notice the comma
t2 = t.Thread(target=squares,args=(numbers,)) #notice the comma
t1.start()
t2.start()

#main thread has to wait for the child thread to complete
t1.join()
t2.join()

endtime = time.time()
print('Total time taken : ',endtime-begintime)
```
